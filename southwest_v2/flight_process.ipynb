```python
import pandas as pd
import numpy as np
```

```python
df_flights = pd.read_csv("sw_flights.csv", header=0)
df_bg = pd.read_csv("airports_background.csv", header=0)
```

```python
for col in df_bg: print (col)
```

```python
# Create map
d_map = {}
for index, row in df_bg.iterrows():
    d_map[row["Code"].strip()] = {
        "name" : row["Display"],
        "city_pop" : row["City_Population"],
        "ap_size" : row["Airport_Size"],
        "annual_passengers" : row["Passengers_Year"],
        "sunny_days" : row["Sunny_Days"]
    }
```

```python
for key in d_map["RNO"]:
    df_flights["depart_" + key] = df_flights["departAp"].apply(lambda x: np.nan if not d_map[x][key] else d_map[x][key])
    df_flights["dest_" + key] = df_flights["destAp"].apply(lambda x: np.nan if not d_map[x][key] else d_map[x][key])
```

```python
df_flights["depart_pop"] = df_flights["departAp"].apply(lambda x: np.nan if not d_map[x]["city_pop"] else d_map[x]["city_pop"])
```

```python
df_flights.sort_values("cost")
```

```python
df_flights.to_csv("agg.csv", index=False)
```
